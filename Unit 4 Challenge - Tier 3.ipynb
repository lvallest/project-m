**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
#Test

Brent1998 = final[(final['Borough'] == Brent) & (final['Year']==1998)]
Brent2018 = final[(final['Borough'] == Brent) & (final['Year']==2018)]

BrentMean1998 = round(Brent1998['Average Price'].mean(), 2)
BrentMean2018 = round(Brent2018['Average Price'].mean(), 2)

print("The ratio between 1998 to 2018 for Brent is " + str(BrentMean1998) + " pounds to " + str(BrentMean2018) + " pounds!!")
```

```python
#final['Average Price'] = final['Average Price'].astype('float', errors = 'ignore')
```

```python
final['Average Price'] = pd.to_numeric(final['Average Price'])
```

```python
final.isna().sum()
```

```python
print(final.dtypes)
```

```python
def price_ratio(p):
    y1998 = float(p.loc[p['Year'] == 1998, 'Average Price'].values[0])
    y2018 = float(p.loc[p['Year'] == 2018, 'Average Price'].values[0])

    if y2018 != 0: 
        ratio = y1998 / y2018
        return ratio
    else:
        return None
```

```python
print(price_ratio(final[final['Borough']=='Brent']))
print(price_ratio(final[final['Borough']=='Camden']))
```

```python
result = {}

for b in final['Borough'].unique():
    borough = final[final['Borough'] == b]
    result[b] = price_ratio(borough)

print(result)
```

```python
ratios = pd.DataFrame(result.items(), columns=['Borough', 'Price Ratio']).set_index('Borough')

ratios.reset_index()
```

```python
ratios_graph = ratios.sort_values(by='Price Ratio')
ratios_graph.head(10)
```

```python
ratios_graph.plot.bar()
plt.xlabel('Borough')
plt.ylabel('Price Ratio')
plt.title('Price Ratio by Burough')
```
